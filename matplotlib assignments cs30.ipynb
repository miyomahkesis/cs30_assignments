```python
# Matplotlib Challenge 1
import matplotlib.pyplot as plt

list = [1,4,7,2,10,2] # List of numbers
plt.plot(list, marker = 'x') # Plots list as a line graph and marks them with an x
plt.show() # Outputs the line graph
```

```python
# Matplotlib Challenge 2
import random
import matplotlib.pyplot as plt 

list = [] # Empty list for random numbers
for x in range(0,6): # Repeats six times
    numbers = random.randint(0,21) # Gets a random integer
    list.append(numbers) # Add those six numbers into the list
plt.pie(list, labels=list) # Plots the list into a pie chart and labels each piece
plt.show() # Outputs the pie chart
```
